##### Librerias

```python
import json
import numpy as np
import pandas as pd
from googletrans import Translator
```

# Ejercicio 1

### Importación de datos

#### Sentimientos

```python
sentimientos = open("Sentimientos.txt") 
valores = {} 
for linea in sentimientos: 
    termino, valor = linea.split("\t") 
    valores[termino] = int(valor) 
```

#### Tweets
Al leer los datos del fichero se hace el filtro de que los tweets tengan el campo "text", que es el asociado al texto del tweet. Para lo cual se recorre el fichero linea a linea y se usa la función .loads() para pasar el json que contiene el tweet en un diccionario, se comprueba si en ese diccionario esta el índice "text" y en caso de estarlo se guarda en otro diccionario que como índice tiene el id del tweet y como valor el texto del tweet

```python
tweets = open("Tweets.txt") 
tweet_dict = {} 

for linea in tweets: 
    b = json.loads(linea)
    if 'text' in b.keys():
        tweet_dict[b['id']] = b['text']
```

#### Esto no funciona en caso de que la palabra del sentimiento este al principio o final del tweet
Para comprobar que el índice del sentimiento esta contenido en el tweet se hace uso de "in", lo cual implica que si aparece la cadena de caracteres detecta el sentimiento, es decir, puede generar errores como el siguiente: el sentimiento "ha" lo detecta en la palabra "have". Para solucionar eso se ha propuesto la adición de espacios vacíos al principio y al final del sentimiento, de forma, que ese error ya no se produce, pero genera otro error, dicho error es que la palabra inicial y la palabra final del tweet no se aceptaran como sentimiento nunca.
Además al hacer uso de "in" en caso de que haya un sentimiento repetido solo se contabilizará una vez.

```python
acum_dict_palabra2b = {} 
for i in tweet_dict:
    acum = 0
    for j in valores:
        if " " + j + " " in tweet_dict[i]:
            acum = acum + valores[j]
    acum_dict_palabra2b[i] = acum
```

### Contamos los sentimientos de cada tweet
Para ello llamamos al traductor de google con cada tweet, por defecto traduce al inglés, que en este caso como los sentimientos están en inglés es el idioma correcto.
Después separamos las palabras del tweet por los espacios con el .split(), con ello podemos comparar palabra por palabra con los sentimientos.
El valor acumulado de los sentimientos se guarda en un diccionario en el que el índice es el id del tweet para después poder guardarlo en un diccionario junto al tweet.
Imprimimos por pantalla el tweet original con el valor de sentimiento asociado.

```python
acum_dict_palabra3b = {}
translator = Translator()
for i in tweet_dict:
    acum = 0
    result = translator.translate(tweet_dict[i])
    split = result.text.split(" ")
    for j in valores:
        for palabra in split:
            if j == palabra:
                acum = acum + valores[j]
    acum_dict_palabra3b[i] = acum
    print(tweet_dict[i], " -> ", acum)
```

### Visualización de los datos
Creamos un dataframe para una visualización más sencilla, así cómo, para tratamiento más sencillo de los datos ya generados, y exportamos el dataframe como .csv

```python
tweet_series = pd.Series(tweet_dict, name='Tweet')
palabra3bucles = pd.Series(acum_dict_palabra3b, name='palabra3bucles')
df = pd.DataFrame(tweet_series)
df["sentimientos"] = pd.Series(palabra3bucles)
df.to_csv("tweet_sentimientos.csv")
```
