##### Librerias

```python
import json
import numpy as np
import pandas as pd
from googletrans import Translator
```

# Ejercicio 2

### Importación de datos

#### Sentimientos

```python
sentimientos = open("Sentimientos.txt") 
valores = {} 
for linea in sentimientos: 
    termino, valor = linea.split("\t") 
    valores[termino] = int(valor) 
```

#### Tweets
Heredado del ejercicio 1


```python
tweets = open("Tweets.txt") 
tweet_dict = {} 

for linea in tweets: 
    b = json.loads(linea)
    if 'text' in b.keys():
        tweet_dict[b['id']] = b['text']
        
acum_dict_palabra3b = {}
translator = Translator()
for i in tweet_dict:
    acum = 0
    result = translator.translate(tweet_dict[i])
    split = result.text.split(" ")
    for j in valores:
        for palabra in split:
            if j == palabra:
                acum = acum + valores[j]
    acum_dict_palabra3b[i] = acum

tweet_series = pd.Series(tweet_dict, name='Tweet')
palabra3bucles = pd.Series(acum_dict_palabra3b, name='palabra3bucles')
df = pd.DataFrame(tweet_series)
df["sentimientos"] = pd.Series(palabra3bucles)
```

#### Opción 1 
Imprimimos el valor total del tweet junto a las palabras que no tienen valor. 'Versión 1, asignando el valor del sentimiento del tweet a las palabras'

```python
translator = Translator()
for i in tweet_dict:
    print("Tweet con id: " + str(i) + "\n{")
    result = translator.translate(tweet_dict[i])
    split = result.text.split(" ")
    for palabra in split:
        bandera = False
        for j in valores:
            if j == palabra:
                bandera = True
        if bandera == False:
            print(palabra + " -> " + str(df["sentimientos"][i]))
        else:
            print(palabra + " -> " + "Sentimiento en el csv original")
    print("}\n")
```

#### Opción 2
Imprimimos el valor medio, valor del sentimiento entre número de palabras, redondeado con 3 decimales

```python
translator = Translator()
for i in tweet_dict:
    print("Tweet con id: " + str(i) + "\n{")
    result = translator.translate(tweet_dict[i])
    split = result.text.split(" ")
    for palabra in split:
        bandera = False
        for j in valores:
            if j == palabra:
                bandera = True
        if bandera == False:
            print(palabra + " -> " + "%.3f"%(df["sentimientos"][i]/len(split)))
        else:
            print(palabra + " -> " + "Sentimiento en el csv original")
    print("}\n")
```

#### Opción 3
Comprobamos si el valor del sentimiento del tweet es 0, en caso afirmativo imprimos que el tweet tiene un sentimiento de 0, para evitar divisiones entre 0, en caso de que el valor del tweet sea distinto de 0 realizamos la versión 2: 'Versión 2, asignando la media del valor del tweet, siendo esta media el número de palabras del tweet dividido entre el valor de sentimiento del tweet'

```python
translator = Translator()
for i in tweet_dict:
    print("Tweet con id: " + str(i) + "\n{")
    if df["sentimientos"][i] != 0:
        result = translator.translate(tweet_dict[i])
        split = result.text.split(" ")
        for palabra in split:
            bandera = False
            for j in valores:
                if j == palabra:
                    bandera = True
            if bandera == False:
                if df["sentimientos"][i] != 0:
                    print(palabra + " -> " + "%.3f"%(len(split)/df["sentimientos"][i]))
                else:
                    print(palabra + " -> " + "")
            else:
                print(palabra + " -> " + "Sentimiento en el csv original")
    else:
        print("El tweet tiene un sentimiento de 0")
    print("}\n")
```
